```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading synthetic data obtained from https://zenodo.org/record/5172126#.Y-bTBxzMJkh for females and changing the values to cm
female_NewModel=pd.read_excel("NewModel/Transformed_female.xlsx")
female_NewModel=female_NewModel.div(10)

#Reading synthetic data obtained from https://zenodo.org/record/5172126#.Y-bTBxzMJkh for males and changing the values to cm
male_NewModel=pd.read_excel("NewModel/Transformed_male.xlsx")
male_NewModel=male_NewModel.div(10)

#Reading data obtained from ANSUR_II_FEMALE_Public for females and changing the values to cm
female_ANSUR2=pd.read_excel("ANSUR2/Transformed_Female.xlsx")
female_ANSUR2=female_ANSUR2.div(10)

#Reading data obtained from ANSUR_II_MALE_Public for males and changing the values to cm
male_ANSUR2=pd.read_excel("ANSUR2/Transformed_Male.xlsx")
male_ANSUR2=male_ANSUR2.div(10)
```

```python
#Descriptive Analysis of selected features for synthetic data obtained from https://zenodo.org/record/5172126#.Y-bTBxzMJkh for females

n_bins = 200
fig, ((ax0, ax1,ax2,ax3), (ax4, ax5,ax6,ax7),(ax8, ax9, ax10,ax11)) = plt.subplots(nrows=3, ncols=4,figsize=(12, 9))
feature=female_NewModel.columns
num=[ax0,ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11]
for f, n in zip(feature, num):
    n.hist(female_NewModel[f], n_bins, histtype='bar', color = "gray")
    n.set_title(f)
    n.axvline(x=female_NewModel[f].mean(), color='navy', label="Mean")
    n.axvline(x=female_NewModel[f].min(), color='navy', label="Min")
    n.axvline(x=female_NewModel[f].mean(), color='orange', label="Mean")
    n.axvline(x=female_NewModel[f].median(), color='black', label="Median")
    n.axvline(x=female_NewModel[f].max(), color='green', label="Max")

fig.tight_layout()
plt.show()


```
